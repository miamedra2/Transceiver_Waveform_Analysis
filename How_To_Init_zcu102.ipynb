#### 2. SI570 MGT clock reprogramming
The Si570 MGT clock gets reprogrammed to 148.5 MHz at time = 5.240061 s.  This is an issue because the transceiver/refclk design expects the MGT clock frequency to be at 153.6 MHz instead.  I haven't figured out how to delete the node from device tree.  So the solution is to use the ZCU102 Board User Interface tool to reprogram during the boot cycle.  Note that you can't reprogram the board after the boot cycle has finished.  

<img src="images/Board_User_Interface_snapshot.png" alt="Drawing" style="width: 500px;">
